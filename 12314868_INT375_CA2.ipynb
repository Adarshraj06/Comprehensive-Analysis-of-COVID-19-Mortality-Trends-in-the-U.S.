```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Provisional_COVID-19_Deaths_by_Sex_and_Age.csv")
df.head()
```

```python
# Dataset Overview
print("Shape of the dataset:", df.shape)

print("\nColumns:\n", df.columns.tolist())

print("\nMetadata (Info):")
df.info()

print("\nSummary Statistics:")
df.describe(include='all')
```

```python
#Data Cleaning
# Replace 'Suppressed' and 'Missing' with NaN
df.replace(['Suppressed', 'Missing'], np.nan, inplace=True)

# Convert numerical columns to proper types
columns_to_convert = ['COVID-19 Deaths', 'Total Deaths', 'Pneumonia Deaths',
                      'Influenza Deaths', 'Pneumonia and COVID-19 Deaths']

for col in columns_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Check missing values
print("Missing Values:")
print(df.isnull().sum())

# Drop rows with essential data missing
df.dropna(subset=['COVID-19 Deaths'], inplace=True)
```

```python
#Data Type Conversion
# Convert 'Month' column to numeric if it exists
if 'MMWR Month' in df.columns:
    df['MMWR Month'] = pd.to_numeric(df['MMWR Month'], errors='coerce')

# Check types
df.dtypes
```

```python
print("Numerical Summary:\n")
print(df[columns_to_convert].describe())

# Histogram
plt.figure(figsize=(8, 4))
sns.histplot(df['COVID-19 Deaths'], kde=True, bins=30,color='royalblue', edgecolor='red')
plt.title('Distribution of COVID-19 Deaths')
plt.xlabel('COVID-19 Deaths')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Sex', y='COVID-19 Deaths', data=df)
plt.title('Outliers in COVID-19 Deaths by Sex')
plt.show()
```

```python
#Correlation Matrix & Heatmap
# Select numerical columns
num_df = df[columns_to_convert].dropna()

# Correlation matrix
corr = num_df.corr()

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix of Death Counts")
plt.show()
```

# Objective 1: COVID-19 Deaths by Age Group & Sex

```python
# Keep only necessary columns for this analysis
df_age_sex = df[['Sex', 'Age Group', 'COVID-19 Deaths']].copy()

# Clean numeric column
df_age_sex['COVID-19 Deaths'] = pd.to_numeric(df_age_sex['COVID-19 Deaths'], errors='coerce')

# Drop missing values
df_age_sex.dropna(inplace=True)
```

```python
# Group by Age Group and Sex
grouped_age_sex = df_age_sex.groupby(['Age Group', 'Sex'])['COVID-19 Deaths'].sum().reset_index()

# Sort age group for proper ordering (optional: custom order if needed)
age_order = ['Under 1 year', '1-4 years', '5-14 years', '15-24 years', '25-34 years', '35-44 years',
             '45-54 years', '55-64 years', '65-74 years', '75-84 years', '85 years and over']
grouped_age_sex['Age Group'] = pd.Categorical(grouped_age_sex['Age Group'], categories=age_order, ordered=True)
grouped_age_sex.sort_values('Age Group', inplace=True)
```

```python
plt.figure(figsize=(16, 9))
sns.barplot(data=grouped_age_sex, x='Age Group', y='COVID-19 Deaths', hue='Sex',hue_order=['Male', 'Female','All Sexes'])
plt.xticks(rotation=45)
plt.title('COVID-19 Deaths by Age Group and Sex')
plt.xlabel('Age Group')
plt.ylabel('Total COVID-19 Deaths')
plt.tight_layout()
plt.show()
```

# Objective 2: Total COVID-19 Deaths by Sex

```python
# Group total COVID-19 deaths by sex
df_sex_total = df[['Sex', 'COVID-19 Deaths']].copy()
df_sex_total['COVID-19 Deaths'] = pd.to_numeric(df_sex_total['COVID-19 Deaths'], errors='coerce')

# Remove nulls
df_sex_total.dropna(inplace=True)

# Group by sex
total_by_sex = df_sex_total.groupby('Sex')['COVID-19 Deaths'].sum().reset_index().sort_values(by='COVID-19 Deaths', ascending=False)

# Display the result
total_by_sex
```

```python
# Bar Chart
plt.figure(figsize=(8, 5))
sns.barplot(data=total_by_sex, x='Sex', y='COVID-19 Deaths', palette='pastel')
plt.title("Total COVID-19 Deaths by Sex")
plt.ylabel("Total Deaths")
plt.xlabel("Sex")
plt.tight_layout()
plt.show()
print("\n")
# Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(total_by_sex['COVID-19 Deaths'], labels=total_by_sex['Sex'],
        autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title("Percentage of COVID-19 Deaths by Sex")
plt.show()
```

# **Objective 3: Comparison of COVID-19, Pneumonia, and Influenza Deaths**

```python
# Select relevant columns
df_compare = df[['COVID-19 Deaths', 'Pneumonia Deaths', 'Influenza Deaths']].copy()

# Convert to numeric
df_compare = df_compare.apply(pd.to_numeric, errors='coerce')

# Drop rows with all NaN values in the three columns
df_compare.dropna(how='all', inplace=True)

# Fill remaining NaN values with 0
df_compare.fillna(0, inplace=True)

# Quick check
df_compare.describe()
```

```python
# Correlation matrix
correlation_matrix = df_compare.corr()
correlation_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Between COVID-19, Pneumonia, and Influenza Deaths')
plt.show()
```

Scatter Plots for Comparison

```python
# 1. COVID-19 vs Pneumonia Deaths
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_compare, x='Pneumonia Deaths', y='COVID-19 Deaths', color='teal')
plt.title("COVID-19 vs Pneumonia Deaths")
plt.xlabel("Pneumonia Deaths")
plt.ylabel("COVID-19 Deaths")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#2.COVID-19 vs Influenza Deaths
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_compare, x='Influenza Deaths', y='COVID-19 Deaths', color='orange')
plt.title("COVID-19 vs Influenza Deaths")
plt.xlabel("Influenza Deaths")
plt.ylabel("COVID-19 Deaths")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Objective 4: Month with Peak COVID-19 Mortality for Each State

```python
# Copy relevant columns
df_peak = df[['State', 'Month', 'COVID-19 Deaths']].copy()

# Drop rows where any of the essential values are missing
df_peak.dropna(subset=['State', 'Month', 'COVID-19 Deaths'], inplace=True)

# Convert 'COVID-19 Deaths' to numeric
df_peak['COVID-19 Deaths'] = pd.to_numeric(df_peak['COVID-19 Deaths'], errors='coerce')

# Convert Month to datetime
df_peak['Month'] = pd.to_datetime(df_peak['Month'], errors='coerce')

# Drop rows with parsing errors
df_peak.dropna(inplace=True)
```

```python
# Get the row with the max deaths per state
peak_months = df_peak.loc[df_peak.groupby('State')['COVID-19 Deaths'].idxmax()].sort_values('COVID-19 Deaths', ascending=False)

# Reset index
peak_months.reset_index(drop=True, inplace=True)

# Display top 10
peak_months.head(10)
```

```python
# Take top 10 states with highest peak deaths
top10_peaks = peak_months.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=top10_peaks, x='COVID-19 Deaths', y='State', palette='magma')
plt.title('Top 10 States with Highest Monthly COVID-19 Death Peaks')
plt.xlabel('COVID-19 Deaths (Peak Month)')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

# Objective 5: COVID Deaths as % of Total Deaths

```python
# Copy relevant columns
df_percent = df[['State', 'Month', 'COVID-19 Deaths', 'Total Deaths']].copy()

# Convert columns to numeric
df_percent['COVID-19 Deaths'] = pd.to_numeric(df_percent['COVID-19 Deaths'], errors='coerce')
df_percent['Total Deaths'] = pd.to_numeric(df_percent['Total Deaths'], errors='coerce')

# Drop rows with missing or zero total deaths (to avoid divide-by-zero)
df_percent.dropna(subset=['COVID-19 Deaths', 'Total Deaths'], inplace=True)
df_percent = df_percent[df_percent['Total Deaths'] > 0]

# Calculate percentage
df_percent['COVID Death %'] = (df_percent['COVID-19 Deaths'] / df_percent['Total Deaths']) * 100
```

```python
# Group by state and calculate mean percentage
state_percent_avg = df_percent.groupby('State')['COVID Death %'].mean().sort_values(ascending=False).reset_index()

# Display top 10 states
state_percent_avg.head(10)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=state_percent_avg.head(10), x='COVID Death %', y='State', palette='coolwarm')
plt.title('Top 10 States by Average COVID Death %')
plt.xlabel('Average % of Total Deaths')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

# Objective 6: Monthly COVID-19 Death Trends

```python
# Extract only required columns
df_trend = df[['Month', 'COVID-19 Deaths']].copy()

# Convert 'Month' to datetime and 'COVID-19 Deaths' to numeric
df_trend['Month'] = pd.to_datetime(df_trend['Month'], errors='coerce')
df_trend['COVID-19 Deaths'] = pd.to_numeric(df_trend['COVID-19 Deaths'], errors='coerce')

# Drop missing values
df_trend.dropna(inplace=True)
```

```python
# Sum COVID deaths per month
monthly_deaths = df_trend.groupby('Month')['COVID-19 Deaths'].sum().reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_deaths, x='Month', y='COVID-19 Deaths', marker='o', color='navy')
plt.title('Monthly COVID-19 Death Trends in the U.S.')
plt.xlabel('Month')
plt.ylabel('Total COVID-19 Deaths')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
